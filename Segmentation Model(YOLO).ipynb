```python
import cv2
from google.colab.patches import cv2_imshow
from ultralytics import YOLO
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"/content/malignant (9).png")
grey_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
resized_img = cv2.resize(grey_img, (225,225))
cv2.imwrite('image.jpg',resized_img)
```

```python
model = YOLO(r"/content/runs/segment/train/weights/best.pt")
```

```python
model.predict(source="/content/image.jpg",save=True,show_labels=True,show_conf=True,conf=0.5,save_txt=False,save_crop=False,line_thickness=1)
```
